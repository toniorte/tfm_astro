```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfCVmags = pd.read_csv('tables/cv_mag.csv')
qsomags = pd.read_csv('tables/qso_mag.csv')
wdmsmags = pd.read_csv('tables/wdms_mag.csv')
obamags = pd.read_csv('tables/oba_mag.csv')
```

```python
def colors(pdaframe):
    pdaframe['J454']=(pdaframe['J0470']+pdaframe['J0520'])/2 - pdaframe['J0490']
    pdaframe['J666']=(pdaframe['J0640']+pdaframe['J0690'])/2 - pdaframe['J0660']
```

```python
colors(dfCVmags)
colors(qsomags)
colors(wdmsmags)
colors(obamags)
```

```python
plt.figure(figsize=(14,8))
plt.scatter(qsomags['J454'], qsomags['J666'], s=2, c='r', label='non-CV')
plt.scatter(wdmsmags['J454'], wdmsmags['J666'], s=2, c='r')
plt.scatter(obamags['J454'], obamags['J666'], s=2, c='r')
plt.scatter(dfCVmags['J454'], dfCVmags['J666'], s=2, c='g', label='CV')
plt.xlabel('(J0470+J0520)/2 - J0490')
plt.ylabel('(J0640+J0690)/2 - J0660')
plt.legend()
```

```python
plt.figure(figsize=(14,8))
plt.scatter(qsomags['J454'], qsomags['J666'], s=2, c='r', label='QSO')
plt.scatter(wdmsmags['J454'], wdmsmags['J666'], s=2, c='b', label='WDMS')
plt.scatter(obamags['J454'], obamags['J666'], s=2, c='y', label='OBA')
plt.xlabel('(J0470+J0520)/2 - J0490')
plt.ylabel('(J0640+J0690)/2 - J0660')
plt.legend()
```
